```python
input_dir ='/content/drive/My Drive/DataFace/'
categories = ['Elon_musk', 'Obama']
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = []
labels = []
```

```python
import os
from skimage.io import imread
from skimage.transform import resize
import numpy as np
from skimage.color import rgb2gray
```

```python
for category_idx, category in enumerate(categories):
    for file in os.listdir(os.path.join(input_dir, category)):
        img_path = os.path.join(input_dir, category, file)
        img = imread(img_path)
        img = resize(img, (15, 15))
        data.append(img.flatten())
        labels.append(category_idx)
```

```python
data = np.asarray(data)
labels = np.asarray(labels)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X, X_test, y, y_test= train_test_split(data, labels, test_size=0.2, random_state=42)
```

```python
#DecisionTree Classifier
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X, y)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
precision = precision_score(y_test, y_pred, average='weighted')
precision
```

```python
recall = recall_score(y_test, y_pred, average='weighted')
recall
```

```python
f1 = f1_score(y_test, y_pred, average='weighted')
f1
```

```python
from sklearn.decomposition import PCA
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
image_data_scaled = scaler.fit_transform(data)
```

```python
pca = PCA()
pca_image_data = pca.fit_transform(image_data_scaled)
```

```python
explained_variance_ratio = pca.explained_variance_ratio_

plt.figure(figsize=(8, 5))
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, alpha=0.7, align='center', color='b')
plt.step(range(1, len(explained_variance_ratio) + 1), np.cumsum(explained_variance_ratio), where='mid', color='r')
plt.ylabel('Explained Variance Ratio')
plt.xlabel('Principal Component')
plt.title('Scree Plot and Cumulative Variance')
plt.grid(True)
plt.show()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans_labels, cmap='viridis', marker='o', edgecolor='k', s=50)
plt.title('K-Means Clustering')
plt.show()
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=10)
dbscan_labels = dbscan.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=dbscan_labels, cmap='plasma', marker='o', edgecolor='k', s=50)
plt.title('DBSCAN Clustering')
plt.show()
```

```python
hierarchical = AgglomerativeClustering(n_clusters=2)
hierarchical_labels = hierarchical.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=hierarchical_labels, cmap='rainbow', marker='o', edgecolor='k', s=50)
plt.title('Hierarchical Clustering')
plt.show()
```
